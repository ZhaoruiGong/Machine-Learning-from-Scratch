## 7.3 PyTorch自动微分
### 7.3.1 反向传播：标量与非标量output

```python
x = torch.arange(4.0, requires_grad  = True)
x, x.grad  # 默认是None
```

```python
# 以y = 2x^T x为例
y = 2*torch.dot(x,x)
y
```

```python
y.backward()
x.grad   # dy/dx = 4x
```

```python
x.grad == 4*x  # 手动检验
```

```python
# 默认torch会累积梯度值，所以在计算其他f(x)对x的梯度前，要先清除之前的值
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
# 非标量变量的反向传播
x.grad.zero_()
y = x*x
y.sum().backward()
# y.backward() 注意直接这样写会报错，因为y是一个向量，无法直接进行反向传播,因为不知道你的优化目标是什么，需要先求和得到标量
x.grad, x
```

```python
# 或者对向量调用backward需要传入一个gradient参数，该参数指定微分函数关于self的梯度。这个方法更通用。
x.grad.zero_()
y = x*x  # 需要再定义一遍，否则上面y.backward()会报错，因为y已经被求和了，不是之前的那个向量了。
y.backward(torch.ones_like(x))  # ∂(y·v)/∂x = ∂(y₁v₁ + y₂v₂ + ... + yₙvₙ)/∂x, where v = torch.ones_like(x)
x.grad, x                       # i.e. y·v = y₁ + y₂ + ... + yₙ = sum(y)，因此二者等价。
```

```python
# 另一个例子
x = torch.tensor([2.0, 3.0], requires_grad=True)
y = x * x

# 假设我们只关心第一个输出
v = torch.tensor([1.0, 0])
y.backward(v)
print(x.grad)  # 输出: [4.0, 0.0] 因为只对第一个输出求导

# 假设我们想给不同输出不同权重
x.grad.zero_()
y = x * x
v = torch.tensor([0.5, 2])
y.backward(v)
print(x.grad)  # 输出: [2.0, 12.0] (因为 0.5*2x₁ 和 2.0*2x₂)
```

### 7.3.2 分离计算

```python
# 希望将某些计算移动到computation graph之外，来节省内存。e.g. x -> y, x, y -> z, 希望将y视作一个常数
x = torch.arange(4, requires_grad=True, dtype=torch.float32)
y = x*x
u = y.detach()  # detach()会返回一个新的tensor，它与原tensor共享数据，但不再是计算图的一部分
z = u*x
v = torch.ones_like(z)
z.backward(v)
x.grad, x.grad == u
```

```python
# 随后我们可以在y上调用反向传播，得到y对x的梯度
x.grad.zero_()
y.backward(v)
x.grad, x.grad == 2*x
```

### 7.3.3 Python控制流下的梯度计算
自动微分的好处包括，即使构建函数的计算图需要通过控制流，如条件、循环和其他函数调用，也不影响自动微分

```python
def f(x):
    y = x * 2
    while y.norm() < 1000:
        y *= 2
    if y.sum() > 0:
        z = y
    else:
        z = 100 * y
    return z

a = torch.randn(size=((3,1)), requires_grad=True)
b = f(a)   # 注意f(a)对a是分段线性的，即对任意a，都存在k s.t. b = k*a, 因此b.backward()不会报错
v = torch.ones_like(b)
b.backward(v)
a.grad, a.grad == torch.abs(b/a)
```

## 7.4 矢量化加速
在模型训练时，当训练数据集n很大，我们经常希望同时处理小批量的样本，因此应进行矢量化从而利用线性代数库，而不是对每个数据点进行算力高昂的for循环。

```python
%matplotlib inline
import math
import time
import numpy as np
import torch
from d2l import torch as d2l
```

```python
n = 10000  # 10000-dim vector
a, b = torch.ones(n), torch.ones(n)

class Timer: #@save
    '''记录多次运行时间，进行时间基准测试'''
    def __init__(self):
        self.times = []
        self.tik = None
        self.start() # 自动启动

    def start(self):
        '''启动计时器'''
        self.tik = time.time()
    
    def stop(self):
        '''停止计时器并将时间记录在列表中'''
        self.times.append(time.time() - self.tik)
        return self.times[-1]

    def avg(self):
        return sum(self.times) / len(self.times)

    def sum(self):
        return sum(self.times)

    def cumsum(self):
        return torch.tensor(self.times).cumsum().tolist()
```

```python
# 用for循环执行每一位加法
c = torch.zeros(n)
timer = Timer()
for i in range(n):
    c[i] = a[i] + b[i]
print(f'{timer.stop():.6f} sec')
```

```python
timer.start()
d = a+b
print(f'{timer.stop():.6f} sec')
# 可见，矢量/tensor化可带来数量级的运算加速。
```
